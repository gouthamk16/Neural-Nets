```python
import tensorflow as tf
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout
```

```python
X,y = make_circles(n_samples=1000, factor=0.5, noise = 0.1)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
X.shape
```

```python
from keras.optimizers import Adam, RMSprop
from keras.losses import BinaryCrossentropy

#Network-1
model1=Sequential()
model1.add(Dense(5, activation='relu', input_shape=(X.shape[1],)))
model1.add(Dense(1, activation='sigmoid'))

model1.compile(optimizer="sgd", loss=BinaryCrossentropy(from_logits=True), metrics=['accuracy'])

history1=model1.fit(X, y, epochs=100, batch_size=16)
```

```python
#Network-2
model2=Sequential()
model2.add(Dense(5, activation='sigmoid', input_shape=(X.shape[1],)))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(optimizer=Adam(learning_rate=0.001), loss=BinaryCrossentropy(from_logits=True), metrics=['accuracy'])

history2=model2.fit(X, y, epochs=100, batch_size=16)
```

```python
#Network-3
model3=Sequential()
model3.add(Dense(3, activation='sigmoid', input_shape=(X.shape[1],)))
model3.add(Dense(3, activation='sigmoid'))
model3.add(Dense(3, activation='sigmoid'))
model3.add(Dense(1, activation='sigmoid'))

model3.compile(optimizer=Adam(learning_rate=0.001), loss=BinaryCrossentropy(from_logits=True), metrics=['accuracy'])

history3=model3.fit(X, y, epochs=100, batch_size=16)
```
